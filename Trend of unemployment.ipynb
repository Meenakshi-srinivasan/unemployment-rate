# Analysis

Question 1 : What are the top 3 states that has most unemployed people(in percent)?

```python
overall_df.head(2)
```

```python
#Calculation of mean value of each row
overall_mean = overall_df.apply(lambda x:x.mean(),axis=1)
overall_mean
```

```python
#Assigning mean values to a new dataframe column
overall_df['mean']=overall_mean
```

```python
#sorting values and fetching the top 3 records
top_3_states = overall_df.sort_values(by='mean',ascending=False).iloc[0:3]
```

```python
#plotting countries with atmost unemployment rate
plt.figure(figsize = (8,8))
plt.pie(top_3_states['mean'],labels=top_3_states.index,colors=['#FD3A5A','#FE8296','#F9B4A2'],explode=(0.1,0.,0),shadow=True,autopct='%1.0f%%')
plt.title('Countries with atmost unemployment rate from 1981-2005 (in %)')
plt.savefig('Countries with atmost unemployment rate.png',dpi=300)
```

```python
#joining top_3_states with female_df
f_top_table = top_3_states.join(female_df,lsuffix='_general',rsuffix='_female')
```

```python
#displayng the dataframe after joining
f_top_table
```

```python
#Joining top_3_states with male_df
overall_top_table = top_3_states.join(male_df,lsuffix='_general',rsuffix='_male')
overall_top_table
```

```python
#plotting the countries with atmost unemployment rate
plt.figure(figsize=(8,5))
plt.plot(f_top_table.mean_general,marker='D',label="Overall unemployed rate")
plt.plot(f_top_table.mean_female,marker='D',label="Female unemployed rate")
plt.plot(overall_top_table.mean_male,marker='D',label="Male unemployed rate")
plt.title('Countries with atmost unemployment rate from 1980-2005 (in %)')
plt.legend()
plt.savefig('Atmost unemployment rate comparison.png',dpi=300)
```

Question : What are the top 3 states with least unemployment rate?

```python
#Sorting the low_3_states and fetching the top 3 records
low_3_states = overall_df.sort_values(by='mean',ascending=True).iloc[0:3]
```

```python
#Displaying the result of low_3_states
low_3_states
```

```python
#plotting the countries with least unemployment rate that shows the composition of each country
plt.figure(figsize=(10,8))
plt.pie(low_3_states['mean'],labels=low_3_states.index,explode=(0,0,0.1),colors=['cyan','green','blue'],shadow=True,autopct='%1.0f%%')
plt.title('Countries with least unemployment rate from 1981 to 2005(in %)')
plt.savefig('Countries with least unemployment rate.png',dpi=300)
```
